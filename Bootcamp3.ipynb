```python
#for문(1)

리스트 = [1,2,3,4]

for i in 리스트:
  제곱 = i ** 2
  print('{}의 제곱은 {}입니다'.format(i,제곱))

## .format은 문자형 안에 {}에 변수를 지정하여 표현하는 방법입니다.
## 즉,문자열 중간 중간에 특정 변수의 값을 넣어주기 위해서 사용하는 것 입니다.
## 우리가 .format()안에 i와 제곱이라는 변수를 지정해줬으므로
## i=1일 때는 '1의 제곱은 1*1입니다.' 출력
## i=2일 때는 '2의 제곱은 2*2입니다.' 출력 and so on..
```

###2.2.2인덱싱

####리스트 인덱싱

```python
#리스트 인덱싱
리스트 = [1,2,3,4,5,6,7,8]

print(리스트[0])      ##파이썬은 0이 첫번째, 1이 두번째를 의미합니다.
print(리스트[-1])     ##-1은 마지막번째를 의미합니다.
print(리스트[2:])     ##세번째부터 끝까지
print(리스트[2:7])    ##세번째부터 일곱번째(여덟번째X)까지
print(리스트[-4:-1])  ##다섯번째부터 일곱번째까지
print(리스트[:])      ##전체 선택하기
```

####위치를 기준으로 데이터 선택하기

```python
#위치를 기준으로 데이터 선택하기(1): index기준
종가.loc['2021-03-23':'2021-03-31']

##loc는 location의 약어로, 사람이 읽을 수 있는 라벨 값으로 특정 값들을 골라오는 방법이라고 생각하면 됩니다.
##리스트인덱싱과 비슷하게 '2021-03-01'인덱스부터 '2021-03-19'인덱스까지를 불러온다는 의미입니다!
```

image.png

```python
#위치를 기준으로 데이터 선택하기(2): column기준
종가.loc[:,'A005930':'A035720']

##.loc[인덱스범위,칼럼범위]
##종가.loc[:,'A000020':'A000050']는 인덱스는 다 불러오고, 칼럼은 'A000020'부터 'A000050'까지 불러오겠다는 의미입니다.
```

image.png

```python
#위치를 기준으로 데이터 선택하기(3): index + column 기준
종가.loc['2021-03-23':'2021-03-31','A005930':'A035720']

##.loc[인덱스범위,칼럼범위]
##자 그럼 종가.loc['2021-03-01':'2021-03-19','A000020':'A000050']는
##인덱스 기준으로 '2021-03-01'부터 '2021-03-19'까지, 컬럼 기준으로 'A000020'부터 'A000050'까지 불러오겠다! 라는 뜻이겠군요.
```

image.png

####순서를 기준으로 데이터 선택하기

```python
#순서를 기준으로 데이터 선택하기(1): index기준
종가.iloc[0:10]

##iloc는 integer location의 약어로, 데이터 프레임의 행이나 칼럼의 순서를 나타내는 정수로 특정 값을 추출해오는 방법입니다.
##리스트인덱싱과 비슷하게 0번째 부터 9번째까지를 불러온다는 의미입니다!
```

image.png

```python
#순서를 기준으로 데이터 선택하기(2): column기준
종가.iloc[:,0:9]

##.iloc[인덱스범위,칼럼범위]
##종가.iloc[:,0:9]는 인덱스는 다 불러오고, 칼럼은 0번째부터 8(9미포함)번째까지 불러오겠다는 의미입니다.
```

image.png

```python
#순서를 기준으로 데이터 선택하기(3): index + column기준
종가.iloc[3:5, 13:17]

##.iloc[인덱스범위,칼럼범위]
##자 그럼 종가.loc[3:5,13:17]은
##인덱스 기준으로 3행부터 5(4아님)행까지, 컬럼 기준으로 13열부터 16(17아님)열까지 불러오겠다! 라는 뜻이겠군요.

##만약에 연속된 범위가 아닌 특정 인덱스들이나 컬럼들을 지정하고 싶으면 리스트로 넘겨주면 됩니다.(그림참고)
## 예시: 종가.iloc[[1,3,5],[[3,11,15]]]
```
