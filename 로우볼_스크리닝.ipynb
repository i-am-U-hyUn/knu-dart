```python
import numpy as np
import pandas as pd
```

```python
profit = pd.read_excel('C:/Users/User/Desktop/다트/2021.2 모의투자/영업이익.xlsx', index_col=0) #3년간 영업이익 >=0
trade = pd.read_excel('C:/Users/User/Desktop/다트/2021.2 모의투자/거래대금.xlsx', index_col=0) #그중 거래대금 상위 60%
day_yield = pd.read_excel('C:/Users/User/Desktop/다트/2021.2 모의투자/일간수익률.xlsx', index_col=0) #그중 변동성 낮은 순으로 30개 종목 선정
```

```python
profit
```

```python
profit = profit.transpose() #transpose() : 행열 뒤바꾸기. 주의사항 : transpose는 한번만!
profit
```

```python
profit = profit[profit[:]>0] #3년 이상 영업이익 >= 0
profit
```

```python
profit.dropna(axis=0, inplace=True) #NaN값(영업이익 0이하) 제거, NaN값이 하나라도 있는 row(index) 제거
profit                               #inplace=True : 기존의 데이터프레임을 건드리겠다
```

```python
profit_screen_index = profit.index #3년이상 영업이익 >= 0인 인덱스 리스트만들기
profit_screen_index
```

```python
trade #거래대금
```

```python
trade = trade.transpose()
trade
```

```python
trade_profit = trade.loc[profit_screen_index] #영업이익>=0인 종목만 대상으로 거래대금 계산
trade_profit
```

```python
trade_profit.sort_values(by='2021-10-08',ascending=False,inplace=True) #거래대금이 큰 순서대로 정렬
trade_profit
```

```python
trade_profit_screen_index = trade_profit.iloc[:int(len(trade_profit.index)*0.6)].index #거래대금 상위 60% 종목 추출
trade_profit_screen_index
```

```python
day_yield = day_yield.transpose()
```

```python
day_yield = day_yield.loc[trade_profit_screen_index]
```

```python
lowvol_list = []
for i in day_yield.index:
    lowvol_list.append(day_yield.loc[i].std()) #각 종목 일간수익률 표준편차(변동성) 리스트 만들기
```

```python
lowvol_list
```

```python
day_yield['volatility'] = lowvol_list #확인용
day_yield
```

```python
lowvol_df = pd.DataFrame(index=day_yield.index) #깔끔하게 하나 DataFrame을 새로 만들어봅시다
lowvol_df
```

```python
lowvol_df['volatility'] = lowvol_list
lowvol_df
```

```python
lowvol_df.sort_values(by='volatility', inplace=True) #변동성이 낮은 순서대로 정렬
lowvol_df
```

```python
lowvol_df = lowvol_df.iloc[:30] #상위 30개 종목 선정
lowvol_df
```
